# Python Scikit Learn

Scikit-learn is a free software machine learning library for the Python programming language. It features various classification, regression and clustering algorithms including support vector machines, random forests, gradient boosting, k-means and decision tree.

## Decision Tree with Python Scikit Learn

A decision tree is a decision support tool that uses a tree-like model of decisions and their possible consequences, including chance event outcomes, resource costs, and utility. It is one way to display an algorithm that only contains conditional control statements.

Importing necesaary libraries for the task:

```python
import sklearn.datasets as datasets
import pandas as pd
```

Let's import the iris dataset. Iris dataset is preinstalled dataset in sklearn library about iris flowers. This is as follows:

```python
iris = datasets.load_iris()
dir(iris)
```

Here, we just need the feature_names column for our task:

```python
iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)

#let's check the first  and last 5 data entries of the dataset
print(iris_df.head())
print(iris_df.tail())
```

```python
y=iris.target
print(y)
```

**We have our required data. Let's use the Decision Tree Algorithm:**

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(iris_df,y)

print('Decision Tree Classifer Created')
```

### Let us visualize the Decision Tree:

```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**Now let us input a new data to Decision Tree algorithm:**

```python
new_data = {'sepal length (cm)':[6.2], 'sepal width (cm)':[3.2], 'petal length (cm)':[4.5], 'petal width (cm)':[1.6]}
```

```python
x_pred = pd.DataFrame(data=new_data)
```

```python
print(x_pred)
```

Let's see the predicted value:

```python
y_pred = dtree.predict(x_pred)
print(y_pred)
```
